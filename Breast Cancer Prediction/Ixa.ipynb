```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

Loading dataset into frame 

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head(10)
```

Adding the target

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head(25)
```

```python
#finding the no of rows and columns in dataset
data_frame.shape
```

```python
# Finding the information about datatypes,column names
data_frame.info()
```

```python
# finding the frequency of target variable
data_frame['label'].value_counts()
```

0 is used for malignant and 1 for benign

```python
# checking for null values
data_frame.isnull().sum()
```

```python
# some statistics
data_frame.describe()
```

```python
data_frame.groupby('label').mean()
```

Separate feature and label

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model_1 = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

model_1.fit(X_train, Y_train)
```

Finding Accuracy

```python
# accuracy on training data
X_train_prediction = model_1.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training dataset is ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model_1.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test dataset is ', test_data_accuracy)
```
